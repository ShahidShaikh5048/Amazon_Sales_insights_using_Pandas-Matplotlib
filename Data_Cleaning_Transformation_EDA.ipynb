```python
def Rename_column_Names(df, old_name, new_name):
    df.rename(columns={old_name: new_name}, inplace=True)
    
Rename_column_Names(new_df, "Item Type", "Item_type")
Rename_column_Names(new_df, "Sales Channel", "Sales_channel")
Rename_column_Names(new_df, "Order Priority", "Order_priority")
Rename_column_Names(new_df, "Order Date", "Order_date")
Rename_column_Names(new_df, "Order ID", "Order_ID")
Rename_column_Names(new_df, "Ship Date", "Ship_date")
Rename_column_Names(new_df, "Units Sold", "Units_sold")
Rename_column_Names(new_df, "Unit Price", "Unit_price")
Rename_column_Names(new_df, "Unit Cost", "Unit_cost")
```

```python
def change_datatype_str_to_datetime(column_name):
    return pd.to_datetime(column_name)
```

```python
new_df["Order_date"] = new_df["Order_date"].apply(change_datatype_str_to_datetime)
new_df["Ship_date"] = new_df["Ship_date"].apply(change_datatype_str_to_datetime)
```

#### Standardizing text fields by converting first character to Uppercase

```python
def standadize_values(value):
    return value.capitalize()

new_df["Region"] = new_df["Region"].apply(standadize_values)
new_df["Country"] = new_df["Country"].apply(standadize_values)
new_df["Item_type"] = new_df["Item_type"].apply(standadize_values)
new_df["Sales_channel"] = new_df["Sales_channel"].apply(standadize_values)
```

#### Formating numerical values(Rounding upto 2 decimal places)

```python
def Rounding_values(value, round_upto):
    return round(value, round_upto)

new_df["Unit_price"].apply(Rounding_values, round_upto=2)
```

```python
new_df["Unit_cost"].apply(Rounding_values, round_upto=2)
```

# Data Transformation

```python
def create_new_column(row, Column1, Column2):
    return row[Column1] * row[Column2] 
```

```python
new_df["Total_revenue"] = new_df.apply(create_new_column, args=("Units_sold", "Unit_price"), axis=1)
```

```python
new_df["Total_cost"] = new_df.apply(create_new_column, args=("Units_sold", "Unit_cost"), axis=1)
```

```python
new_df["Total_profit"] = new_df["Total_revenue"] - new_df["Total_cost"]
```
